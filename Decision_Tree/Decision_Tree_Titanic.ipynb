```python
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train_titanic.csv")
df.head()
```

```python
inputs = df.drop(['Name','Ticket','Cabin'], axis = 'columns')
target = df['Survived']
inputs.head()
```

```python
#Age and Embarked has missing values
inputs['Age'] = inputs['Age'].fillna(inputs['Age'].mean())
inputs['Embarked'] = inputs['Embarked'].fillna('S')
```

```python
Sex = LabelEncoder()
Embarked = LabelEncoder()
```

```python
inputs['Sex_n'] = Sex.fit_transform(inputs['Sex'])
inputs['Emb_n']= Embarked.fit_transform(inputs['Embarked'])
inputs.head()
```

```python
inputs.drop(['Sex','Embarked'], axis = 'columns', inplace = True)
inputs.head()
```

```python
inputs.drop(['Survived'], axis = 'columns', inplace = True )
```

```python
inputs.head()
```

```python
features_tarin,features_test,target_tarin,target_test = train_test_split(inputs,target, test_size = 0.2)
print(features_tarin.head())
print(target_test.head())
```

```python
clf = tree.DecisionTreeClassifier(min_samples_split= 100)
```

```python
clf.fit(features_tarin, target_tarin)
```

```python
y_pred = clf.predict(features_test)
```

```python
#Now measuring accuracy of model
acc = accuracy_score(target_test,y_pred)
print("Accuracy of our model is : {}".format(acc))
```
