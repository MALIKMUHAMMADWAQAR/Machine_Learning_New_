# Using trignometric function sin,cos,tan in NUmpy

import numpy as np
import matplotlib.pyplot as plt


```python
arr_x = np.arange(0,10,0.2)
print(arr_x)
arr_y= np.sin(arr_x)
```

```python
#Now ploting values of sin function
plt.plot(arr_x, arr_y)
plt.show()
```

```python
#plotiing for cos function
arr_y= np.cos(arr_x)
plt.plot(arr_x, arr_y)
```

```python
#plotting for tan function
arr_y= np.tan(arr_x)
plt.plot(arr_x, arr_y)
```

# Random sampling with Numpy

```python
import random
#random function always give values greater than 0 and less than 1
np.random.random(5)
```

```python
#Generate a 3x3 random number array (values between 0 and 1)
np.random.random((3,3))
```

```python
#Creating matrix of 3x3 having numbers greater than 1 and less than 5
np.random.randint(1,5,(3,3))
```

```python
# creating a 3d array
# random.randint(low, high=None, size=None, dtype=int)
#Return random integers from the “discrete uniform” distribution of the specified dtype in the “half-open” interval [low, high).
#If high is None (the default), then results are from [0, low)
np.random.randint(10,20,(3,3,3))
```

```python
# randn create a random 1d array and value will be between zero and 1
print(np.random.randn(10))
```

```python
#choice function in random library gives the random elements of your choice from given sequence

arr_1 = np.arange(0,9)
for i in range(10):
    print(np.random.choice(arr_1))
```

```python
#Permutation
arr_2 = np.arange(0,5)
for i in range(10):
    print(np.random.permutation(arr_2))
```
