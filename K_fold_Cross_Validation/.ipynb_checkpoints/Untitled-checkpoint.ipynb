```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
target = df['label']
inputs = df.drop(['label'], axis = 'columns')
```

```python
kf = KFold(n_splits=5)

def get_score(x_train,x_test,y_train,y_test,model):
    model.fit(x_train,y_train)
    score = model.score(x_test,y_test)
    return score
```

```python
svm_scores = []
log_scores = []
forest_scores = []

model_1 = SVC()
model_2 = LogisticRegression()
model_3 = RandomForestClassifier()

for train_index, test_index in kf.split(inputs):
    features_train, features_test = inputs.loc[train_index], inputs.loc[test_index] #To access values using index we use loc 
    labels_train, labels_test = target.loc[train_index], target.loc[test_index]
    
    svm_scores.append(get_score(features_train, features_test, labels_train, labels_test, model_1))
    log_scores.append(get_score(features_train, features_test, labels_train, labels_test, model_2))
    forest_scores.append(get_score(features_train, features_test, labels_train, labels_test, model_3))
    
   
    
```

```python
print("SVM has a score of : {}".format(svm_scores))
print("log_scores has a score of : {}".format(svm_scores))
print("forest_scores has a score of : {}".format(svm_scores))
```
